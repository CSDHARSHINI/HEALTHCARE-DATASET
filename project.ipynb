```python
df = df.drop(['Insurance Provider', 'Billing Amount', 'Room Number'], axis=1, errors='ignore')
```

## DATA VISUALIZATION FOR THE DATASET

```python
plt.hist(df['Admission Type'], bins=10, color='darkred', edgecolor='black')
plt.title('Histogram of patients condition')
plt.xlabel('condition of the patients')
plt.ylabel('levels')
plt.grid(True)
plt.show()
```

BOXPLOT FOR THE DATASET

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Medical Condition', y='Admission Type', data=df,hue='Medical Condition', dodge=False)
plt.title('Box Plot of Admission Type by Medical Condition')
plt.xlabel('Medical Conditiony')
plt.ylabel('Admission Type')
plt.xticks(rotation=45)
plt.legend([], frameon=False)
plt.show()
```

#SCATTERPLOT FOR DATASET

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Medication', y='Test Results', data=df)
plt.title('Scatter Plot for patients condition ')
plt.xlabel('Medication')
plt.ylabel('Test Results')
plt.show()
```

```python
sns.countplot(x='Blood Type', data=df)
plt.title('Bar Plot of Blood Type')
plt.show()
```

#BOXPLOT FOR THE DATASET

```python
plt.figure(figsize=(5, 6))
sns.boxplot(x='Admission Type', y='Test Results', data=df)
plt.title('Box Plot of Test Results by Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Test Results')
plt.show()
```

LINE PLOT FOR THE DATASET

```python
plt.plot(df['Gender'], df['Admission Type'], marker='o', linestyle='-')
plt.title('Line Plot')
plt.xlabel('Gender')
plt.ylabel('Admission Type')
plt.grid(True)
plt.show()
```
